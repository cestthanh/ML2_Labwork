#ENCODING DATASET

Before applying PCA, we need to encode the categorical features and standardize the
data. PCA relies on variance to identify the principal components.

For Student habits performance dataset:

• Gender: One-Hot Encoding (Convert ”Female” and ”Male” and ”Other”)

• Part-Time Job: Label Encoding (No = 0, Yes = 1)

• Diet Quality: Ordinal Encoding (Poor < Fair < Good) ⇒ Poor = 0, Fair = 1,
Good = 2

• Parental Education Level: Ordinal Encoding (None < High School < Bachelor <
Master) ⇒ None = 0, High School = 1, Bachelor = 2, Master = 3

• Internet Quality: Ordinal Encoding (Poor < Average < Good) ⇒ Poor = 0,
Average = 1, Good = 2

• Extracurricular Participation: Label Encoding (No = 0, Yes = 1)

```python

from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder

# Initialize encoders
label_encoder = LabelEncoder()
ordinal_encoder = OrdinalEncoder()
onehot_encoder = OneHotEncoder(sparse_output=False)

# 1. Gender: One-Hot Encoding (including 'Other' category)
gender_encoded = onehot_encoder.fit_transform(df[['gender']])

# Create a DataFrame with One-Hot Encoded gender columns
gender_encoded_df = pd.DataFrame(gender_encoded, columns=onehot_encoder.get_feature_names_out(['gender']))

# Concatenate this back to the original dataframe, dropping the original 'gender' column
df_encoded = pd.concat([df.drop(columns=['gender']), gender_encoded_df], axis=1)

# 2. Part-Time Job: Binary Encoding (No = 0, Yes = 1)
df_encoded['part_time_job'] = label_encoder.fit_transform(df_encoded['part_time_job'])

# 3. Diet Quality: Corrected Ordinal Encoding (Poor = 0, Fair = 1, Good = 2)
diet_quality_order = ['Poor', 'Fair', 'Good']
ordinal_encoder = OrdinalEncoder(categories=[diet_quality_order])
df_encoded['diet_quality'] = ordinal_encoder.fit_transform(df_encoded[['diet_quality']])

# 4. Parental Education Level: Corrected Ordinal Encoding (None = 0, High School = 1, Bachelor = 2, Master = 3)

# Fill missing values in 'parental_education_level' with 'None'
df['parental_education_level'] = df['parental_education_level'].fillna('None')

education_order = ['None', 'High School', 'Bachelor', 'Master']
ordinal_encoder = OrdinalEncoder(categories=[education_order])
df_encoded['parental_education_level'] = ordinal_encoder.fit_transform(df_encoded[['parental_education_level']])

# 5. Internet Quality: Corrected Ordinal Encoding (Poor = 0, Average = 1, Good = 2)
internet_quality_order = ['Poor', 'Average', 'Good']
ordinal_encoder = OrdinalEncoder(categories=[internet_quality_order])
df_encoded['internet_quality'] = ordinal_encoder.fit_transform(df_encoded[['internet_quality']])

# 6. Extracurricular Participation: Binary Encoding (No = 0, Yes = 1)
df_encoded['extracurricular_participation'] = label_encoder.fit_transform(df_encoded['extracurricular_participation'])

# Display the updated dataset with correct encoding applied
df_encoded.head()

```

```python
# Save data
df_encoded.to_csv(PATH+'/student_habits_performance_encoding.csv',index=False)
```
