```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Customer Churn_cleaning.csv")

df.info()
df.describe()
```

| Bước preprocessing            | Mục đích chính                           |
| ----------------------------- | ---------------------------------------- |
| Tách nhãn                     | Phân tách dữ liệu đầu vào và nhãn        |
| Xử lý thiếu dữ liệu           | Giữ dữ liệu sạch, không lỗi Null         |
| Mã hóa categorical            | Chuyển biến dạng số để mô hình hiểu được |
| Chuẩn hóa dữ liệu             | Đưa dữ liệu cùng thang đo                |
| Xử lý mất cân bằng (nếu có)   | Cân bằng dữ liệu để mô hình không bias   |
| Tách train-test               | Đánh giá mô hình trên dữ liệu chưa học   |
| Trích xuất/lựa chọn đặc trưng | Giảm chiều hoặc chọn biến quan trọng     |


BỎ CỘT LABELS

```python
X = df.drop(columns=['churn'])

X.head()
```

Chuyển hai biến tariff_plan và status từ giá trị 1/2 về 0/1

```python
df_encoded = X.copy()

df_encoded['tariff_plan'] = df_encoded['tariff_plan'].apply(lambda x: 0 if x == 1 else 1)
df_encoded['status'] = df_encoded['status'].apply(lambda x: 0 if x == 1 else 1)

df_encoded.head()
```

```python
df_encoded.to_csv("Customer Churn_encoding.csv", index = False)
```

Chuẩn hóa dữ liệu (Standardization): Giúp các biến có thang đo đồng nhất, tránh biến có giá trị lớn chi phối mô hình.

```python
scaler = StandardScaler()

scaler.fit(df_encoded)
scale_data = scaler.transform(df_encoded)

df_scaled = pd.DataFrame(scale_data, columns=df_encoded.columns)

df_scaled.head()
```

```python
df_scaled.to_csv("Customer Churn_scaled.csv", index = False)
```
